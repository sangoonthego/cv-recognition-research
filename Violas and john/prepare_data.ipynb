```python
import os
import cv2
import numpy as np
```

```python
def rec(x_center,y_center,w,h,W,H): 
     x_center *= W
     y_center *= H
     w*=W
     h*=H
     xmin = int(x_center - w/2)
     ymin = int(y_center - h/2)
     xmax = int(x_center + w/2)
     ymax = int(y_center + h/2)
     return [xmin,ymin,xmax,ymax]
def creative_data(dir_train,dir_label,folder_save,max_number_negative): 
    count=1
    file_images = []
    for root,dirs,file in os.walk(dir_train):
        file_images.append(file)
    file_labels = [] 
    for root,dirs,file in os.walk(dir_label): 
        file_labels.append(file)
    file_images = np.array(file_images)
    file_labels = np.array(file_labels)
    loc_humans = []
    for file_txt in file_labels[0]:
        path_txt=os.path.join(dir_label,file_txt)
        with open(path_txt,"r") as f: 
            loc_human = []
            for line in f:
                data=line.strip().split()
                loc_human.append( list(map(float, data[1:])))
        loc_humans.append(loc_human)
    for i, file_img in enumerate(file_images[0]): 
        print(f"đang tiến hành :{(count/len(file_images[0]))*100:.2f} %")
        path_img = os.path.join(dir_train,file_img)
        img = cv2.imread(path_img)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        H,W = gray.shape
        # tìm các khung có chứa mặt người để cắt  
        rec_human=[]
        for j in loc_humans[i]:   
            rec_human.append(rec(j[0],j[1],j[2],j[3],W,H))
        #cắt và lưu vào folder positive nếu tồn tại 
        pos_dir = os.path.join(folder_save, "positive")
        os.makedirs(pos_dir, exist_ok=True)
        for k, (xmin, ymin, xmax, ymax) in enumerate(rec_human):
            if xmax <= xmin or ymax <= ymin:
                continue
            face_crop = gray[ymin:ymax, xmin:xmax]
            face_crop=cv2.resize(face_crop,(24,24))
            
            if face_crop.size == 0:
                continue
            save_path = os.path.join(pos_dir, f"{i}_{k}.jpg")
            cv2.imwrite(save_path, face_crop)
        # tìm các ảnh không chứa mặt người và lưu vào negative
        neg_dir = os.path.join(folder_save, "negative")
        os.makedirs(neg_dir, exist_ok=True)
        for n in range(max_number_negative):
            x = np.random.randint(0, W - 24)
            y = np.random.randint(0, H - 24)
            box_neg = [x, y, x+24, y+24]
            overlap=False 
            for (xmin, ymin, xmax, ymax) in rec_human:
                if not (box_neg[2] < xmin or box_neg[0] > xmax or
                        box_neg[3] < ymin or box_neg[1] > ymax):
                    overlap = True
                    break
            if overlap==True: 
                continue 
            neg_crop = gray[y:y+24, x:x+24]
            if neg_crop.size == 0:
                continue
            save_path = os.path.join(neg_dir, f"{i}_{n}.jpg")
            cv2.imwrite(save_path, neg_crop)
        count+=1
```
